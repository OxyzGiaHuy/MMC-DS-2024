# Bài 1: Chữ số La Mã [link](https://leetcode.com/problems/roman-to-integer/description/)
Chữ số La Mã được biểu thị bằng bảy ký hiệu khác nhau: I, V, X, L, C, D và M.  
image.png  
Ví dụ: 2 được viết là II bằng chữ số La Mã, chỉ cần hai số cộng lại với nhau. 12 được viết là XII, đơn giản là X + II. Số 27 viết là XXVII tức là XX + V + II.

Chữ số La Mã thường được viết từ lớn nhất đến nhỏ nhất từ trái sang phải. Tuy nhiên, chữ số của bốn không phải là IIII. Thay vào đó, số bốn được viết là IV. Bởi vì số một đứng trước số năm nên chúng ta trừ nó thành bốn. Nguyên tắc tương tự áp dụng cho số chín, được viết là IX. Có sáu trường hợp sử dụng phép trừ:  

I có thể được đặt trước V (5) và X (10) để tạo thành 4 và 9.  
X có thể được đặt trước L (50) và C (100) để tạo thành 40 và 90.  
C có thể được đặt trước D (500) và M (1000) để tạo thành 400 và 900.  
Cho một số La Mã, hãy chuyển nó thành số nguyên.    
```
Example 1:
Input: s = "III"
Output: 3
Explanation: III = 3.

Example 2:
Input: s = "LVIII"
Output: 58
Explanation: L = 50, V= 5, III = 3.

Example 3:
Input: s = "MCMXCIV"
Output: 1994
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.
```

````python
def romanToInt(s):
    romanDict = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }
    num = 0
    for i in range(0, len(s)-1):
        num += romanDict[s[i]]
        if romanDict[s[i]] < romanDict[s[i+1]]:
            num -= 2*romanDict[s[i]]
    num += romanDict[s[len(s) - 1]]
    return num

print(romanToInt("MCMXCIV"))
````

# Bài 2: Tìm số FA [link](https://leetcode.com/problems/single-number/description/)  
Cho một mảng số nguyên không trống, mỗi phần tử xuất hiện hai lần ngoại trừ một phần tử. Tìm cái duy nhất đó.

Bạn phải triển khai giải pháp có độ phức tạp thời gian chạy tuyến tính và chỉ sử dụng không gian bổ sung không đổi.
```
Example 1:
Input: nums = [2,2,1]
Output: 1

Example 2:
Input: nums = [4,1,2,1,2]
Output: 4

Example 3:
Input: nums = [1]
Output: 1
```

````python
def singleNumber(nums):
    single_num = 0;
    for num in nums:
        single_num ^= num;
    return single_num

print(singleNumber([1,4,2,2,1]))
````

# Bài 3: Substring dài nhất [link](https://leetcode.com/problems/longest-common-prefix/description/)  
Viết hàm tìm chuỗi tiền tố chung dài nhất trong một mảng các chuỗi.  
Nếu không có tiền tố chung, trả về một chuỗi trống " ".  
```
Example 1:
Input: strs = ["flower","flow","flight"]
Output: "fl"

Example 2:
Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.
```

````python
def longestCommonPrefix(strs):
    ans = ""
    strs = sorted(strs)
    firstString = strs[0]
    lastString = strs[-1]
    minLen = min(len(firstString), len(lastString))
    for i in range(minLen):
        if firstString[i] < lastString[i]:
            break
        ans += firstString[i]
    return ans
    
longestCommonPrefix(["flower","flow","flight"])
````

# Bài 4: Top các phần tử k thường gặp [Link](https://leetcode.com/problems/top-k-frequent-elements/description/)
Cho một mảng số nguyên nums và một số nguyên k, trả về k phần tử thường xuyên xuất hiện nhất. Bạn có thể trả lời câu trả lời theo bất kỳ thứ tự nào.
```
Example 1:
Input: nums = [1,1,1,2,2,3], k = 2
Output: [1,2]

Example 2:
Input: nums = [1], k = 1
Output: [1]
```

````python
def topKFrequent(nums, k):
    frequency_dict = {}
    for num in nums:
        frequency_dict[num] = frequency_dict.get(num, 0) + 1
    frequency_dict = sorted(frequency_dict.items(), key=lambda x: x[1], reverse=True)

    return [x[0] for x in frequency_dict[:k]]

print(topKFrequent([1,1,1,2,2,3], 2))
````
