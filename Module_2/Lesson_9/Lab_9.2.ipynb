Bây giờ, dựa vào thông tin được cung cấp từ câu lệnh trên, chúng ta cần kiểm tra xem các cột có đúng định dạng chưa và đưa những cột dữ liệu về đúng dạng, điều này giúp cho việc phân tích dữ liệu được hiệu quả.
Có  nhiều cách khác nhau để xử lý về kiểu dữ liệu, tuy nhiên hãy xử lý một cách có trình tự, đừng chọn cột một cách ngẫu nhiên, điều đó sẽ làm rối quá trình xử lý dữ liệu, cách tốt nhất là hãy xử lý chúng theo một trình tự nhất định. Trình tự đó có thể là từng cột trong dữ liệu, hoặc từng kiểu dữ liệu trong Pandas.
Trong bài asm này, chúng ta sẽ thực hiện bằng cách quan sát từng kiểu dữ liệu theo thứ tự kiểu dữ liệu có sẵn của output áp cuối trong câu lệnh trên: bool, float, int, object.

```python
# Yêu cầu: Hãy in ra những cột dữ liệu có kiểu dữ liệu là bool.
condition = df_google_playstore.dtypes == "bool"
cols_ = condition.index[condition == True].to_list()
df_google_playstore.loc[:, cols_]
```

Có vẻ như chúng ta không gặp khó khăn khi kiểm tra những cột dữ liệu có kiểu bool như trên. Tiếp theo hãy in ra những cột có kiểu dữ liệu là float.

```python
# Yêu cầu: Hãy in ra những cột dữ liệu có kiểu dữ liệu là float.
condition = df_google_playstore.dtypes == "float"
cols_ = condition.index[condition == True].to_list()
df_google_playstore.loc[:, cols_]
```

```python
# Yêu cầu: Hãy chuyển kiểu dữ liệu của Rating Count về dạng int
df_google_playstore["Rating Count"] = df_google_playstore["Rating Count"].astype("int")
```

Lỗi thông báo: Cannot convert non-finite values (NA or inf) to integer. Tức là bạn không thể chuyển thành kiểu dữ liệu int do trong cột dữ liệu của bạn đang có giá trị null.
Có nhiều cách khác nhau để xử lý về dữ liệu null:
- Bỏ những dòng dữ liệu có giá trị null.
- Thay thế giá trị null bằng giá trị khác.

Việc thay thế giá trị null bằng giá trị khác cũng có nhiều cách khác nhau:

- Thay thế bằng giá trị âm (áp dụng khi dữ liệu phân phối dương)
- Thay thế bằng giá trị ngoài phân phối: 9999, 99999, ...
- Thay thế bằng giá trị mean, median, mode (áp dụng khi giả định rằng dữ liệu bị khuyết có thể thuộc phân phối chung của dữ liệu)

```python
# Let's fix it (later because it might change our dataframe )
```

```python
# Yêu cầu: Hãy in ra giá trị lớn nhất và nhỏ nhất của cột Rating Count
print("Min rating count:", df_google_playstore["Rating Count"].min())
print("Max rating count:", df_google_playstore["Rating Count"].max())
```

Từ kết quả trên cho thấy Rating Count thấp nhất là 0, cũng có nghĩa là có một số App nào đó chưa có người đánh giá. Điều này cho thấy tất cả giá trị của cột dữ liệu này mang phân phối dương, do đó để thay thế giá trị cho null, chúng ta chỉ cần gán 1 giá trị âm bất kì (chủ yếu là -1)

```python
# Yêu cầu: Hãy gán giá trị -1 cho giá trị bị khuyết
df_google_playstore["Rating Count"] = df_google_playstore["Rating Count"].fillna(-1)

# Yêu cầu: Hãy chuyển cột Rating Count về kiểu dữ liệu int
df_google_playstore["Rating Count"] = df_google_playstore["Rating Count"].astype(
    "int32"
)

print("Rating Count dtype:", df_google_playstore["Rating Count"].dtypes)
```

Bây giờ, bạn cần chuyển kiểu dữ liệu của Mininum Installs về kiểu dữ liệu int nhưng bằng các
h xóa giá trị Null

```python
# Yêu cầu: Hãy xóa những dòng dữ liệu có giá trị Null hoặc NA ở cột Minimum Installs
# Dropna is more tricky. don't use the same as above, this need to modify the whole
# df_google_playstore with subset like below
df_google_playstore.dropna(subset=["Minimum Installs"], inplace=True)
# Yêu cầu: Hãy chuyển cột Rating Count về kiểu dữ liệu int
df_google_playstore["Minimum Installs"] = df_google_playstore[
    "Minimum Installs"
].astype("int32")

print("Minimum Installs dtype:", df_google_playstore["Minimum Installs"].dtypes)
```

Tiếp tục quan sát các cột với kiểu dữ liệu Object

```python
# Yêu cầu: Quan sát các cột dữ liệu có kiểu dữ liệu là Object
condition = df_google_playstore.dtypes == "object"
condition = condition.index[condition == True].to_list()
df_google_playstore[condition]
```

Ngay sau khi in ra các cột dữ liệu có kiểu object, chúng ta dễ dàng nhận ra 3 cột dữ liệu đang có sai định dạng về thời gian. Cụ thể cột Released, Last Updated, Scraped Time phải có kiểu dữ liệu là Datetime. Do đó chúng ta cần chuyển dữ liệu của 3 cột này về Datetime.
Với phương thức astype, bạn vẫn có thể chuyển dữ liệu về dạng datetime bằng cú pháp .astype('datetime64[ns]'). Tuy nhiên, trong bài học này,chúng ta sẽ sử dụng một phương pháp như sau: 
<b>df['col_datetime'] = pd.to_datetime(df['col_datetime'])<b>

```python
# Yêu cầu: Hãy chuyển đổi 3 cột Released, Last Updated,Scraped Time về dạng Datetime
df_google_playstore["Released"] = pd.to_datetime(df_google_playstore["Released"])
df_google_playstore["Last Updated"] = pd.to_datetime(
    df_google_playstore["Last Updated"]
)
df_google_playstore["Scraped Time"] = pd.to_datetime(
    df_google_playstore["Scraped Time"]
)
```

Quá trình chuyển đổi kiểu dữ liệu sang datetime có thể mất vài phút. 
Sau khi chuyển đổi thành công hãy in ra các cột có dữ liệu là datetime.

```python
# Yêu cầu: Quan sát các cột dữ liệu có kiểu dữ liệu là Datetime
condition = df_google_playstore.dtypes == "datetime64[ns]"
condition = condition.index[condition == True].to_list()
df_google_playstore[condition]
```

```python
# Yêu cầu: Quan sát các cột dữ liệu có kiểu dữ liệu là Object
condition = df_google_playstore.dtypes == "object"
condition = condition.index[condition == True].to_list()
df_google_playstore[condition]
```

Tiếp theo chúng ta hãy xem xét trong dữ liệu này có cột nào là cột có kiểu dữ liệu hạng mục không. Về cơ bản dữ liệu hạng mục cũng có kiểu dữ liệu là object, tuy nhiên trong một cột số lượng duy nhất ít hơn rất nhiều so với tổng số mẫu trong dữ liệu, thường được sử dụng để phân thành các nhóm khác nhau cho từng mẫu dữ liệu. Ví dụ:
- Giới tính: Nam - Nữ.
- Thành phố: TP.HCM - Hà Nội - Huế

Bây giờ để xác định đâu là cột dữ liệu hạng mục, chúng ta cần tính tỉ lệ phần trăm phần tử duy nhất so với số lượng dữ liệu.

```python
# Yêu cầu: Điền vào đoạn code còn thiếu
object_cols = df_google_playstore.select_dtypes("object").columns.values
object_cols = object_cols[1:]
# For this case, we want to see if the percentage of the unique vs len of the columns
# The reason for this is because we want to see the re-occurance of the elements inside a
# columns. For this reason, we want to strike for anything that's have a reasonably low percentage
# of unique, because it can tell the re-occurance rate of a column
per_reoccur = []
for col in object_cols:
    a = df_google_playstore[col]
    b = len(a.unique()) / len(a) * 100
    print(col, ":", b, "%")
    # store the result
    per_reoccur.append(b)
```

Chúng ta có thể thấy rằng với App Name, Với tỉ lệ gần 100% thì có thể dễ dàng khẳng định, đây không phải là dữ liệu hạng mục. Điều này là chính xác vì tên App không liên quan tới việc phân nhóm, mỗi App có một tên khác nhau.

Bạn có thể nhận có khá nhiều cột dữ liệu có tỉ lệ giữa số lượng giá trị duy nhất và số lượng dữ liệu rất nhỏ, điều đó có nghĩa là có những giá trị được lặp lại nhiều lần và điều đó cho chúng thấy đây là dữ liệu hạng mục.

```python
col_category = []
i = 0
# Yêu cầu: Thêm tên cột vào list rỗng ở trên nếu tỉ lệ phần trăm phần tử duy nhất so với số lượng dữ liệu nhỏ hơn 5%
for percent in per_reoccur:
    if percent <= 0.05:
        print(object_cols[i])
        col_category.append(object_cols[i])
    i += 1
```

```python
# Yêu cầu: Chuyển kiểu dữ liệu của các cột được xác định là hạng mục về kiểu Category
for col in col_category:
    df_google_playstore[col] = df_google_playstore[col].astype("category")
```

Có khá nhiều loại kí tự đuôi để hỗ trợ cho giá trị cột Size, chúng ta cũng nên quan sát xem có những loại đuôi nào bằng phương thức Apply kết hợp với lambda và regex

```python
import re


#     if pd.isnull(row):
#     return None
#     match = re.findall(r'[A-Za-z]+',row)
# return match[0] if match else None
# Yêu cầu: Hãy xây dựng hàm để trả về kí tự trong chuỗi
# This is a tricky one if you think about it. I RECOMMEND ALWAYS USE TYPE HINT FOR THIS!
def get_character(row: pd.Series) -> list:
    if pd.isnull(row):
        return None
    match = re.findall(r"[A-Za-z]+", row)
    return match[0] if match else None
    # characters = re.findall(r"[^\d]+", str(row))
    # return characters
    # return row.str.extract()
```
