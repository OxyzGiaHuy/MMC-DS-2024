```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
path = "./DATA/giraffes.jpg"
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display_img(img)
```

**TASK:Apply a binary threshold onto the image.**

```python
img_gray = cv2.imread(path, 0)
_, img_bin = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
display_img(img_gray, cmap='gray')
display_img(img_bin, cmap='gray')
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
img_hsv = cv2.imread(path)
img_hsv = cv2.cvtColor(img_hsv, cv2.COLOR_BGR2HSV)
display_img(img_hsv, cmap ='hsv')
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones((4, 4), dtype=np.float32) * 0.1
kernel
```

```python
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blurred_img = cv2.filter2D(img, - 1, kernel)
# -1 means we keep original number of channels
display_img(blurred_img)
```
